```python
# Create a new column, pre_tip_amount
df['pre_tip_amount'] = (
    df
    [['fare_amount', 'extra', 'mta_tax', 'tolls_amount', 
      'improvement_surcharge', 'congestion_surcharge']]
    .sum(axis='columns')
)
```

```python
# Create a new column, tip_percentage, showing the percent of the fare_amount that people tipped
df['tip_percentage'] = df['tip_amount'] / df['pre_tip_amount']
```

```python
df.columns
```

```python
# What was the overall tip percentage?
df['tip_percentage'].mean()
```

```python
# How many times did people tip more than 100%?
(df['tip_percentage'] > 1).value_counts()
```

```python
# What percent of taxi riders give no tip at all?
(df['tip_percentage'] == 0).value_counts(normalize=True)
```

32.1% of riders gave no tip
67.9% gave some tip

```python
# On which day of the week do people tip the greatest percentage?
# Mon = 0, Tues = 1, Wed = 2, Thu = 3, Fri = 4, Sat = 5, Sun = 6
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'], errors = 'coerce')
df.groupby(df['tpep_pickup_datetime'].dt.day_of_week)['tip_percentage'].mean().sort_values(ascending=False)
```

```python
# At which hour do people tip the greatest percentage?
df.groupby(df['tpep_pickup_datetime'].dt.hour)['tip_percentage'].mean().sort_values(ascending=False)
```

```python
# Do people tip more, on average, in January or July?
df.groupby(df['tpep_pickup_datetime'].dt.month)['tip_percentage'].mean().sort_values(ascending=False)
```

```python
# What was the 1-day period in our data set when people tipped the greatest percentage?
df = df.set_index('tpep_pickup_datetime')
df.resample('1D')['tip_percentage'].mean().sort_values(ascending=False).head(10)
```

```python
# Keep only January and July. First count how many rides fall outside those
# two months (needed for Question 2), before we drop them.
in_jan = (df.index >= '2019-01-01') & (df.index < '2019-02-01')
in_jul = (df.index >= '2019-07-01') & (df.index < '2019-08-01')
rides_outside_jan_jul = int((~(in_jan | in_jul)).sum())

df = df.sort_index()
df = pd.concat([
    df[(df.index >= '2019-01-01') & (df.index < '2019-02-01')],
    df[(df.index >= '2019-07-01') & (df.index < '2019-08-01')]
])

df.resample('1D')['tip_percentage'].mean().sort_values(ascending=False).head(10)
```

```python
df.resample('1ME')['tip_percentage'].mean().dropna()
```

## Analysis 15% - 20 %

Analysis for when people tip their taxi drivers more generously with the customary tip between 15% and 20%.  

```python
df = df[df.index.notna()]
df = df[df['tip_percentage'].notna()]

# Filter only tips between 15% and 20%
custom_tip_df = df[df['tip_percentage'].between(0.15, 0.20)]
```

```python
total_rides = len(df)
custom_tip_rides = len(custom_tip_df)

custom_tip_percentage = custom_tip_rides / total_rides * 100

print("Total rides:", total_rides)
print("Rides with 15%–20% tips:", custom_tip_rides)
print(f"Percentage of rides with 15%–20% tips: {custom_tip_percentage:.2f}%")
```

# **Group by Type:**

```python
#hour
tips_by_hour = custom_tip_df.groupby(custom_tip_df.index.hour)['tip_percentage'].count()
#day
tips_by_day = custom_tip_df.groupby(custom_tip_df.index.day_name())['tip_percentage'].count()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
tips_by_day = tips_by_day.reindex(day_order)
#month
tips_by_month = custom_tip_df.resample('1ME')['tip_percentage'].count()
tips_by_month.index = tips_by_month.index.strftime('%B %Y')
```
